lets extract the names with a regex

```python
import re

def extract_fun_name(input_string):
    
    # The regex pattern
    pattern = r'def\s+([^\s]+)\('
    # Using re.search() to find the first match
    match = re.search(pattern, input_string)
    # Checking if a match was found
    if match:
        # Extracting the matched group, which is the function name in this case
        return match.group(1)
    else: return 'None'
        
extract_fun_name("Here is an example: def exampleFunction( and some more text.")
```

```python
llama_generated['llama_name'] = llama_generated['generated'].apply(extract_fun_name)
```

```python

def extract_function_code_with_regex(code):
    pattern = re.compile(r'\bdef\s+([a-zA-Z_][a-zA-Z0-9_]*)\s*\((.*?)\)\s*:\s*(.*?)(?=def|$)', re.DOTALL)
    matches = pattern.findall(code)
    functions = {}

    for match in matches:
        function_name = match[0]
        function_args = match[1]
        function_body = match[2]
        functions[function_name] = f"def {function_name}({function_args}):\n    {function_body}"

    return functions
```

```python
import json
dataset['llama_prompt'] = ""
dataset['llama_code'] = ""
dataset['llama_names_dict'] = ""
for i in range(dataset.shape[0]):
  line = dataset.loc[i]
  name =   str(line['prompt'])
  p = str(line['prompt'])
  c = str(line['code'])
  d = {}
  for func_name, func in extract_function_code_with_regex(line['prompt']).items():
    if func_name not in names_dict.keys():
        names_dict[func_name] = func_name
    if func_name in c:
      d[func_name] = names_dict[func_name]
      p = p.replace(func_name+"(", names_dict[func_name]+"(" )
      c = c.replace(func_name+"(", names_dict[func_name]+"(" )
  for func_name, func in extract_function_code_with_regex(str(line['code'])).items():
      if func_name not in names_dict.keys():
        d[func_name] = names_dict[func_name]
        names_dict[func_name] = func_name
      d[func_name] = names_dict[func_name]  
      c = c.replace(func_name+"(", names_dict[func_name]+"(")
  dataset.at[i, 'llama_prompt'] = p
  dataset.at[i, 'llama_code'] = c
  dataset.at[i,'llama_names_dict'] = json.dumps(d)
```
